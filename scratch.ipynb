```python
def find_fraction(a,b, reverse=True):
    rat = a/b
    denom, err = sorted([(d, abs(int(rat*d) - rat*d)/d) for d in range(1,20)], key=lambda x: x[1])[0]
    num = int(rat * denom)
    if reverse:
        n2, d2, e2 = find_fraction(b,a,False)
        if e2 < err:
            num = d2
            denom = n2
            err = e2
    g = gcd(denom, num)
    denom /= g
    num /= g
    return int(rat*denom), denom, err
def gcd(a,b):
    if b==0:
        return a
    else:
        return gcd(b, a%b)
```

```python
import numpy as np
irrs = [ (a**(1/b),a,b) for a in range(2,20) for b in range(2,4)]
```

```python
fracs = sorted([ (find_fraction(a[0],b[0]),a[1:],b[1:]) for a in irrs for b in irrs if a[0]/b[0] % 1 != 0 and b[0]/a[0] %1 != 0 ], key=lambda x: x[0][2])
```
